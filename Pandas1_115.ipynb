```python
import pandas as pd
#empty Dataframe
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','john','Smith','William'])
id_series=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id_series}
result=pd.DataFrame(frame)
print("\nSeries to DataFrame\n")
print(result)
```

```python
print("\nExtracting on column:\n")
print(result['Emp'])
```

```python
print("\nAdding new column:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
print("\nDeleting one column:\n")
del result['Age']
print(result)
```

```python
print("\nExtracting the third row:\n")
print(result.loc[2])
```

```python
print("\nSlice rows:\n",result[1:3])
```

```python
d2=pd.DataFrame([['Date',123],['Mark',143]],columns=['Emp','ID'])
print("\nAdding new row values:\n",pd.concat([result,d2]))
```

```python
print("\nDeleting particular row :\n",result.drop(1))
```
