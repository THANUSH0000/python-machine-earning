```python
#open a text file in 'Write' mode to write the given line
f=open("/content/textfile.text",'w')
f.write("first line\n")
f.write("second line\n")
f.write("thrid line\n")
f.close()
#open a text file in 'read' mode to read and print the available line - readline function
k=open("/content/textfile.text",'r')
print(k.readline())
print(k.readline())
print(k.readline())
```

```python
#Open a text file in 'Append' mode to add/append new line inputs

m=open("textfile.txt", 'a')

m.write("fourth line\n")

m.write("fifth line\n")

m.close()

#Open a text file in 'Read' mode to read and print the available full text - read function

p=open("textfile.txt", 'r')

print(p.read())

p.close()
```
