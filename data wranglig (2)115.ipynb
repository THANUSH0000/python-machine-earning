```python
import pandas as pd
#Assign data
data={'Name':['Sai','Yuvasri','Poovannan',
'Sabharish','Sangamithra','Rithika','Rubasri'],
'Age':[20,18,19,20,18,17,19],
'Gender':['M','F','M','M','F','F','F'],
'Marks':[90,76,'NaN',74,65,'NaN',71]}
#convert into datafram
df=pd.DataFrame(data)
#display data
print("Original DataFrame:\n",df)

# calculate average of numeric marks
total_numeric_marks = 0
count_numeric_marks = 0
for ele in df['Marks']:
    # Check if the element is an integer or float
    if isinstance(ele, (int, float)):
        total_numeric_marks += ele
        count_numeric_marks += 1

# Calculate the average after iterating through all elements
avg = 0
if count_numeric_marks > 0:
    avg = total_numeric_marks / count_numeric_marks

#Replace missing values
# The 'NaN' is a string in the input data, so we replace the string 'NaN'.
df=df.replace(to_replace="NaN",value=avg)
#Display data
print("\nReplacing NaN with average marks:\n",df)
```

```python
#Filter top scoring students
df=df[df['Marks']>=75]

#Display data
print("\nFiltering and Dropping func:\n",df)
```
